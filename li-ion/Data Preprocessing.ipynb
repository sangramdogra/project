```python
def red_dat_re(data):
    '''Function to remove if there exist simultaneous profile'''
    ind = []
    length = len(data[0][0][0][0])
                 
    for i in range(length):
        if i == 0:
            ind.append(i)
        elif data[0][0][0][0][i][0][0] != data[0][0][0][0][i-1][0][0]:
            ind.append(i)
            
                    
####################################################
    cycle = []
    for i in ind:
        cycle.append(data[0][0][0][0][i][0][0])
    for i in zip(list(range(len(cycle))), ind):
        if cycle[i[0]] == 'impedance' and i[0]< length-1 and i[0]> 0:
            if cycle[i[0]-1] == cycle[i[0]+1]:
                ind.pop(i[0]-1)
#################################################
    cycle1 = []
    for i in ind:
        cycle1.append(data[0][0][0][0][i][0][0])
        
    if cycle1[0] != 'impedance':
        if cycle1[0] == cycle1[len(cycle1)-1]:
            ind.pop(len(cycle1)-1)
            cycle1.pop(len(cycle1)-1)
    else:
        if cycle1[1] == cycle1[len(cycle1)-1]:
            ind.pop(len(cycle1)-1)
            cycle1.pop(len(cycle1)-1)
    
        
#     print(cycle1.count('charge'))
#     print(cycle1.count('discharge'))
#     print(cycle1.count('impedance'))
#     print(cycle1)
    return ind
```

```python
def c2d(x):
    ''' Function to obtain datetime from dataset'''
    return str(int(x[0])) + '-' + str(int(x[1])) + '-' + str(int(x[2])) + ' ' + str(x[3]) + ':' + str(x[4]) + ':' + str(x[5])
```

```python
def dataframe(X, dig):
    ''' Coversion of cycles to data frame'''
    cycle = []
    temp = []
    date = []
    datarray = []
    for i in red_dat_re(X):
        cycle.append(X[0][0][0][0][i][0][0])
        temp.append(X[0][0][0][0][i][1][0][0])
        date.append(pd.to_datetime(c2d(X[0][0][0][0][i][2][0])))
        datarray.append(X[0][0][0][0][i][3][0])
    charge, tempc, datec, datarrayc  = [], [], [], []
    discharge, tempd, dated, datarrayd   = [], [], [], []
    impedence, tempi, datei, datarrayi = [], [], [], []
    j, k, l = 0, 0, 0
    for i in range(len(cycle)):
        if cycle[i] == 'charge':
            charge.append(str(dig) + '_' + cycle[i] + '_' + str(j))
            j+=1
            tempc.append(temp[i])
            datec.append(date[i])
            datarrayc.append(datarray[i][0])

        elif cycle[i] == 'discharge':
            discharge.append(str(dig) + '_' + cycle[i] + '_' + str(k))
            k+=1
            tempd.append(temp[i])
            dated.append(date[i])
            datarrayd.append(datarray[i][0])
        else:
            impedence.append(str(dig) + '_' + cycle[i] + '_' + str(l))
            l+=1
            tempi.append(temp[i])
            datei.append(date[i])
            datarrayi.append(datarray[i][0])

    df = pd.DataFrame(np.array([charge,tempc,datec, datarrayc]).T, columns = ['cycle', 'temp', 'date', 'data'])
    df['voltage_measured'] = df['data'].apply(lambda x: x[0][0])
    df['current_measured'] = df['data'].apply(lambda x: x[1][0])
    df['temperature_measured'] = df['data'].apply(lambda x: x[2][0])
    df['current_charge'] = df['data'].apply(lambda x: x[3][0])
    df['voltage_charge'] = df['data'].apply(lambda x: x[4][0])
    df['time'] = df['data'].apply(lambda x: x[5][0])
    df.drop(columns = 'data', inplace = True)

    df1 = pd.DataFrame(np.array([discharge,tempd,dated, datarrayd]).T, columns = ['cycle', 'temp', 'date', 'data'])
    df1['voltage_measured'] = df1['data'].apply(lambda x: x[0][0])
    df1['current_measured'] = df1['data'].apply(lambda x: x[1][0])
    df1['temperature_measured'] = df1['data'].apply(lambda x: x[2][0])
    df1['current_charge'] = df1['data'].apply(lambda x: x[3][0])
    df1['voltage_charge'] = df1['data'].apply(lambda x: x[4][0])
    df1['time'] = df1['data'].apply(lambda x: x[5][0])
    df1['capacity'] = df1['data'].apply(lambda x: x[6][0])
    df1.drop(columns = 'data', inplace = True)

    df2 = pd.DataFrame(np.array([impedence,tempi,datei, datarrayi]).T, columns = ['cycle', 'temp', 'date', 'data'])
    df2['sense_current'] = df2['data'].apply(lambda x: x[0][0])
    df2['battery_current'] = df2['data'].apply(lambda x: x[1][0])
    df2['current_ratio'] = df2['data'].apply(lambda x: x[2][0])
    df2['battery_impedence'] = df2['data'].apply(lambda x: x[3][0])
    df2['rectified_impedence'] = df2['data'].apply(lambda x: x[4][0])
    df2['re'] = df2['data'].apply(lambda x: x[5][0])
    df2['rct'] = df2['data'].apply(lambda x: x[6][0])
    df2.drop(columns = 'data', inplace = True)

    
    return df, df1, df2
```

```python
# Posing problem as classification problem
def classification(x):
    ''' Conversion capacity to class'''
    cls = []
    rem = len(x)//4
    for i in range(len(x)):
        if i<= rem:
            cls.append(0)
        elif i> rem and i<= 2*rem:
            cls.append(1)
        elif i> 2*rem and i<= 3*rem:
            cls.append(2)
        else:
            cls.append(3)
    return cls
```

```python
# There are nan values present in data which needs to be removed
def nan_pop(row):
    row = list(row)
    cnt = 0
    for ind, ele in enumerate(row):
        if str(ele) == 'nan':
            row[ind] = 'nan'
            cnt+=1
    for i in range(cnt):
        row.remove('nan')
    return np.array(row)
```

```python
train_charge_data['voltage_measured'] = train_charge_data['voltage_measured'].apply(nan_pop)
train_charge_data['current_measured'] = train_charge_data['current_measured'].apply(nan_pop)
train_charge_data['temperature_measured'] = train_charge_data['temperature_measured'].apply(nan_pop)
```

```python
def get_padding_len(data):
    vector_len = []
    for row in data.iterrows():
        vector_len.append(len(row[1]['voltage_measured']))
    return max(vector_len), vector_len

padding_length, _ = get_padding_len(train_charge_data)
```

```python
# We will be doing pre-padding of 0
def padding(column):
    l = len(column)
    if l < padding_length:
        return np.array([0] * (padding_length-l) + list(column))
    else:
        return column
len(padding(train_charge_data.voltage_charge[0]))
```

```python
def avg_sam(x, feature):
    sampling = []
    for i in range(len(x)):
        sample = []
        rem = len(x[feature][i])//10
        ini = 0
        fin = rem
        if len(x[feature][i]) > 10:
            for j in range(10):
                sample.append(np.nanmean(x[feature][i][ini:fin]))
                ini = ini + rem 
                fin = fin + rem
            sampling.append(sample)
            
    return sampling        
```

```python
voltage_sampling = pd.DataFrame(avg_sam(train_charge_data, 'voltage_measured'))
temperature_sampling = pd.DataFrame(avg_sam(train_charge_data, 'temperature_measured'))
current_sampling = pd.DataFrame(avg_sam(train_charge_data, 'current_measured'))

preprocessed_train_charge_data = pd.concat([train_charge_data[['cycle', 'temp']],voltage_sampling, temperature_sampling,\
                                      current_sampling, train_charge_data['soh'], train_charge_data['class']], axis =1)
preprocessed_train_charge_data.columns = ['cycle', 'temp', 'v1', 'v2', 'v3', 'v4', 'v5', 'v6', 'v7', 'v8', 'v9', 'v10',\
                't1', 't2', 't3', 't4', 't5', 't6', 't7', 't8', 't9', 't10',\
                'c1', 'c2', 'c3', 'c4', 'c5', 'c6', 'c7', 'c8', 'c9', 'c10', 'soh', 'class']

if not os.path.exists('preprocessed_train_charge_data.csv'):
    preprocessed_train_charge_data.to_csv('preprocessed_train_charge_data.csv')
preprocessed_train_charge_data.head()
```
